```python
class InventoryManagement():
    def __init__(self):
        self.inventory = {}
    def add_inventory(self):
        item = int(input("enter number of products"))
        for i in range(item):
            name = input("enter name of product")
            price = int(input("enter price"))
            quantity = int(input("enter Quantity"))
            self.inventory[name] = {
                "price" : price,
                "quantity" : quantity,
                "total_price" : price * quantity}
    def update_inventory(self,product_name):
            if product_name in self.inventory:
                new_name = input("enter new name of product")
                new_price = int(input("update price"))
                new_quantity = int(input("update Quantity"))
                self.inventory[new_name] = {
                    "price" : new_price,
                    "quantity" : new_quantity,
                    "total_price" : new_price * new_quantity}
                if new_name != product_name:
                    del self.inventory[product_name]
            else:
                print("your items in not in the list")
    def display(self):
        for key,value in self.inventory.items():
            print(key)
            for e,g in value.items():
                print(e,g)
        
```

```python
im = InventoryManagement()
```

```python
im.add_inventory()
```

```python
im.display()
```

```python
im.update_inventory("book")
```

```python
im.display()
```
