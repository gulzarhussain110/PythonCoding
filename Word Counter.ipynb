```python
para = input("Enter a paragraph: ")
```

```python
word = para.split()
counters = {}
for i in word:
    if i in counters:
        counters[i] +=1
    else:
        counters[i] = 1
print(counters)
```

```python
for key,value in counters.items():
    if value > 3:
        print(key,value)
```
