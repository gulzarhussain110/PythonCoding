```python
class Liberary:
    def __init__(self):
        self.books = {}
        self.users = {}
        self.issued_books = {}
    def add_user(self):
        name = input("enter your name")
        password = input("enter your password")
        if not self.users:
            self.users[name] = {
                "password" : password,
                "role": "Liberarian"
            }
        elif name in self.users:
            print("username should be unique")            
        else:
            self.users[name]={
                "password" : password,
                "role" : "student"
            }
    def login(self):
        name = input("Enter your username")
        password = input("enter your password")
        if name in self.users:
            if self.users[name]["password"] == password:
                if self.users[name]["role"] == "student":
                    while True:
                        print("1:Search Book \n2:Borrow Book \n3:Return Book \n4:My Booking \n5:Logout")
                        option = int(input("Enter from menu"))
                        if option == 1:
                            self.search_book()
                        elif option ==2:
                            self.borrow_books(name)
                        elif option ==3:
                            self.return_book(name)
                        elif option == 4:
                            self.my_booking(name)
                        elif option == 5:
                            break
                        else:
                            print("you have selected wrong option")
                else:
                    while True:
                        print("1:Add Book \n2:Remove Book \n3:View Book \n4:view_bookers \n5:Logout")
                        option = int(input("Enter from menu"))
                        if option == 1:
                            self.add_books()
                        elif option ==2:
                            self.remove_book()
                        elif option ==3:
                            self.view_books()
                        elif option == 4:
                            self.view_bookers()
                        elif option == 5:
                            break
                        else:
                            print("you have selected wrong option")
            else:
                print("Sorry Wrong Password")
        else:
            print("There is no user with this name")
                        
    def add_books(self):
        book_name = input("Enter the name of book")
        no_of_copies = int(input("enter a number of copies"))
        author = input("Name of author")
        self.books[book_name] = {
            "copies" : no_of_copies,
            "author" : author
        }
    def view_books(self):
        if not self.books:
            print("There is no book in the record")
        else:
            for book,copies in self.books.items():
                print(f"\033[01m{book}\033[0m")
                for author,no_copy in copies.items():
                    print(author,no_copy)
    def search_book(self):
        book = input("Enter the name of book")
        if book in self.books:
            print(f"\033[01m{book}\033[0m")
            for name,copies in self.books[book].items():
                print(name,copies)
        else:
            print("book is not availabe")
    def remove_book(self):
        book = input("enter the book name you want to delete")
        if book in self.books:
            del self.books[book]
            print(f"The book {book} has been deleted")
        else:
            print("There is no book with this name")
    def borrow_books(self,name):
        book = input("enter the name of book")
        if book in self.books:
            if self.books[book]["copies"] > 0:
                print(f"Congratulation Mr {name} you have Booked {book}")
                self.books[book]["copies"] -=1
                if name in self.issued_books:
                    self.issued_books[name].append(book)
                else:
                    self.issued_books[name] = [book]
            else:
                print("Sorry We have no more copies")
        else:
            print("Sorry there is no book with this name")
    def return_book(self,name):
        book = input("Enter the name of Book you wish to return")
        if name in self.issued_books:
            if book in self.issued_books[name]:
                self.books[book]["copies"] +=1
                self.issued_books[name].remove(book)
                print("Congraulation you data has been updated")
            else:
                print(f"you have not borrow {book}")
        
        else:
            print("You have not any book with this name")
    def my_booking(self,name):
        if name in self.issued_books:
            print(self.issued_books[name])
        else:
            print("you don't have any book at your name")
    def view_bookers(self):
        if not self.issued_books:
            print("No one has booked any books")
        else:
            for name,books in self.issued_books.items():
                print(f"\033[01m{name}\033[0m")
                for i in books:
                    print(i)              
```
