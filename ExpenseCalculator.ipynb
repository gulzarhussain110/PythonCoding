```python
expense = []
for i in range(7):
    category = input("Category Name:")
    amount = int(input("Amount: "))
    expense.append(dict(category = category , amount = amount)) 
## you can define dictionary like this as well
## expense.append({"category": category , "amount": amount})
```

```python
print(expense)
```

```python
Total_Expense = 0
for items in expense:
    print(items['category'], items ["amount"])
    Total_Expense += items["amount"]
print("Total Expense:", Total_Expense)
```

```python
temp = 0
ExpensiveCategory = ""
for items in expense:
    if items["amount"] > temp:
        temp = items["amount"]
        ExpensiveCategory = items["category"]
print(f"Expensive category is {ExpensiveCategory}, with amount of {temp}")
    
```
