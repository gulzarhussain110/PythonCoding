```python
students = {}
std_num = int(input("Enter number of Students"))
for i in range(std_num):
    name = input("enter the name of Studnets: ")
    students[name] = {
        "maths":int(input("enter the marks of maths")),
         "Physics":int(input("enter the marks of Physics")),
         "chemistry":int(input("enter the marks of Chemistry")),
         "Bio":int(input("enter the marks of Bio")),
         "Computer":int(input("enter the marks of Computer"))}
```

```python
print(students)
```
