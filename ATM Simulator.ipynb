```python
class acc_():
    def __init__(self):
        self.account = {}
        self.trans = {}
    def creat_user(self):
        user = input("Enter the name of user")
        if user in self.account:
            print("The username is already taken please try new one")
            return
        else:
            pin = int(input("please select the 4 digit pin for your account"))
            amount = int(input("please enter the inital deposit amount"))
            self.account[user] = {
                "pin" : pin,
                "amount" : amount
            }
            self.trans[user]=[f"opening_Balance {amount}"]
    def login(self):
        attempt = 3
        while attempt > 0:
            user_ver = input("enter the user name")
            pin_ver = int(input("enter the pin"))
            flg = False
            if user_ver in self.account:
                flg = True
                if self.account[user_ver]["pin"] == pin_ver:
                    print(f"hallo Mr {user_ver} welcome to your acc")
                    print("Please Select from the Menu\n","1. Check Balance","2. Deposit Money","3. Withdraw Money","4. Transaction History","5. Exit")
                    while True:
                        menu = int(input("Please enter the number form the menu: "))
                        if menu == 1:
                            self.check_balance(user_ver)
                        elif menu == 2:
                            self.deposit_money(user_ver)
                        elif menu == 3:
                            self.Withdraw(user_ver)
                        elif menu == 4:
                            self.Transaction(user_ver)
                        elif menu == 5:
                            break
                        else:
                            print("You have entered the wrong number")
                    break
                else:
                    print("incorrect password")
                    attempt -=1
            if flg == False:
                print("user name is incorrect")
                attempt -=1

    def check_balance(self,user_ver):
        print(f"Mr. {user_ver} your account balance is ",self.account[user_ver]["amount"])
        
    def deposit_money(self,user_ver):
        deposit_amount = int(input("Enter the amout you want to deposit in your account"))
        self.account[user_ver]['amount'] += deposit_amount
        self.trans[user_ver].append(f"amount_deposit {deposit_amount}")
        print(f"Thanks for your deposit. Now your balance of account is \033[01m{self.account[user_ver]["amount"]}\033[0m")
    
    def Transaction(self,user_ver):
        if user_ver in self.trans:
            for names in self.trans[user_ver]:
                print(names)
        else:
            print("The User has no transaction records")
        
    def Withdraw(self,user_ver):
        withdraw_ = int(input("Please enter the amount you want to withdraw"))
        if withdraw_ > self.account[user_ver]['amount']:
            print("\033[01mAmount insufficient\033[0m")
        else:
            self.account[user_ver]["amount"] -= withdraw_
            self.trans[user_ver].append(f"withdraw amount {withdraw_}")
            print(f"you have successfully withdraw the amount {withdraw_}, Now your balance is \033[01m{self.account[user_ver]["amount"]}\033[0m")
    def Exit(self):
        print("Thank you for your visit")
        
```
