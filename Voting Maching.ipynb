```python
candidateList = {
    "iqbal" : 0,
    "Aslam":0,
"vrushab":0
}
```

```python
for i in range(5):
    candidateVote = input("Enter the name of candidate you want to vote")
    if  candidateVote in candidateList:
        candidateList[candidateVote] +=1
    else:
        print(f"There is no candidate with the Name of: {candidateVote}")    
```

```python
temp= 0
Winner = ""
for key,value in candidateList.items():
    if value > temp:
        temp = value
        Winner  = key
print(f"Congratulation {Winner} get the heighest votes with {temp} no of votes")
```
