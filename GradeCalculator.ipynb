```python
no_of_students = int(input("total number of Students"))
students = {}
```

```python
for i in range(no_of_students):
    name = input("Enter the name of Students")
    number = int(input(f"enter number of {name}"))
    students[name]= number
```

```python
print(students)
```

```python
print(students)
```

```python
for key,value in students.items():
    if value <= 50:
        print(f"{key} is fail")
    elif 50 < value <= 60 :
        print(f"Congratualtion! {key} has passed C Grade")
    elif 60 < value <= 70:
        print(f"Congratualtion! {key} has passed B Grade")
    elif 70 < value <= 80:
        print(f"Congratualtion! {key} has passed A Grade")
    else:
        print(f"Congratualtion! {key} has passed A++ Grade")
        
```

```python
temp = 0
Highest_scorer = ""
Grade = 0
for key,value in students.items():
    if value > temp:
        temp = value
        Highest_scorer = key
        Grade = temp
print(f"Highest Scorer is {Highest_scorer} with {Grade}%")
    
    
```
