```python
class Seat_reservation:
    def __init__(self):
        self.row = int(input("rows"))
        self.No_of_seats_in_row = int(input("no of seats"))
        self.bus = {}
        for row in range(0, self.row):
            row_char = chr(65 + row)
            for seat in range(1, self.No_of_seats_in_row+1):
                self.bus[f"{row_char}{seat}"] = None
    def menu(self):
        while True:
            print("\n")
            print("1: Show Seats \n2: Book Seats\n3: Cancel Booking\n4: View_booking\n5: Exit")
            choice = int(input("Please Select from the menu"))
            if choice == 1:
                self.show_seats()
            elif choice == 2:
                self.seat_booking()
            elif choice == 3:
                self.cancel_booking()
            elif choice == 4:
                self.view_booking()
            elif choice == 5:
                break
            else:
                print("you have enter wrong number")
    def seat_booking(self):
        print("\n")
        passenger = input("enter your name")
        print(list(self.bus.keys()))
        seat_ = input("enter your seat number")
        seat_available = seat_.upper()
        if seat_available in self.bus:
            if self.bus[seat_available] == None:
                self.bus[seat_available] = passenger
                print(f"Congratulation Mr. {passenger} you have booked {seat_available} seat")
            else:
                print("Sorry the seat is reserved")
        else:
            print("there is no seat with this number")
    def show_seats(self):
        print("\n")
        for seat,passenger in self.bus.items():
            if passenger is not None:
                print(seat , "='booked'")
            else:
                print(seat , "='Availabe'")
    def cancel_booking(self):
        print("\n")
        name = input("Enter your name")
        seat = input("Enter your seat number")
        if seat in self.bus:
            if self.bus[seat] == name:
                self.bus[seat] = None
                print("Booking cancel successfully")
            else:
                print("this seat is not with this name")
        else:
            print("you have enter the wrong seat number")
    def view_booking(self):
        print("\n")
        flg = False
        for seat,passenger in self.bus.items():
            if passenger is not None:
                print(f"{seat} ==> {passenger}")
                flg = True
        if flg is False:
            print("There is no booking in this bus")
```
