```python
class contact_book():
    def __init__(self):
        self.contact_details = {}
    def menu(self):
        while True:
            print("\n")
            print("please select from menu")
            print("1: Add Contact\n2: View Contact\n3: Search Contact\n4: Update Contact \n5: Delete Contact \n6: Exit")
            option = int(input("Please Enter from menu"))
            if option==1:
                self.add_contact()
            elif option==2:
                self.view_contact()
            elif option==3:
                self.search_contact()
            elif option ==4:
                self.update_contact()
            elif option ==5:
                self.delete_contact()
            elif option ==6:
                break
            else:
                print("you have enter the wrong option")
            
            
    def add_contact(self):
        print("\n")
        name = input("Enter your name")
        if name in self.contact_details:
            print("Name is already been taken. Name should be unique")
        else:
            num = int(input("Enter your contact Number"))
            email = input("Enter your email address")
            
            self.contact_details[name]= dict(number = num, mail = email)
    def view_contact(self):
        print("\n")
        for name,detials in self.contact_details.items():
            print(f"\033[01m{name}\033[0m")
            for num,mail in detials.items():
                print(num,mail)
    def search_contact(self):
        print("\n")
        name = input("enter the name contact you are looking for")
        if name in self.contact_details:
            for key,value in self.contact_details[name].items():
                print(f"\033[01m{key}\033[0m = {value}")
        else:
            print("there is not any contact with this name")
    def update_contact(self):
        print("\n")
        name = input("enter the name you want to update")
        if name in self.contact_details:
            print("1:Number \n2:Email\n3:Both")
            option = int(input("what you want to update"))
            if option ==1:
                new_num = int(input("Please enter the new number"))
                self.contact_details[name]['number']=new_num
                print("Congratulation the phone number has been updated")
            elif option ==2:
                new_mail = input("enter the new mail")
                self.contact_details[name]["mail"]=new_mail
                print("Congratulation the Email has been updated")
            elif option==3:
                new_num = int(input("Please enter the new number"))
                new_mail = input("enter the new mail")
                self.contact_details[name]['number']=new_num
                self.contact_details[name]["mail"]=new_mail
                print("Congratulation your Email and number has been updated")
            else:
                print("Sorry you have selected the wrong number")
        else:
            print("There is no contact with this name")
    def delete_contact(self):
        print("\n")
        name = input("enter the contact you want to delete")
        if name in self.contact_details:
            del self.contact_details[name]
            print("The contact has been remove")
        else:
            print("There is no contact with this name")
        
        
```

```python
contact = contact_book()
```
